# **3. Data Cleaning**

## **3.1 Cleaning the 3 telecom dataframes**

```python
# Cleaning the first telecom dataframe
# First we drop the columns which are not useful in our data analysis
df1.drop('CELL_ON_SITE', axis = 1, inplace = True)
df1.drop('COUNTRY_A', axis = 1, inplace = True)
df1.drop('COUNTRY_B', axis = 1, inplace = True)

# Next we make sure all columns are spelled correctly as in the description dataframe
df1.columns.values[0:1] = ['PRODUCT']
df1.columns.values[2:3] = ['DATE_TIME']

# Afterwards, we go ahead and check if we have any null or missing values
# then sum them so as to decide how to manipulate them
df1.isnull().values.any()
df1.isnull().sum()

# We realize that the column 'SITE_ID' has 716 values
# We substitute the null values with 'Not identified'
df1 = df1.fillna('Not identified')

# We also check for duplicate values so as to drop them
df1.duplicated()
df1.drop_duplicates() 

# Now we load our dataframe
df1
```

```python
# Cleaning the second telecom dataframe
# First we drop the columns we won't use in our analysis
df2.drop('CELL_ON_SITE', axis = 1, inplace = True)
df2.drop('COUNTRY_A', axis = 1, inplace = True)
df2.drop('COUNTRY_B', axis = 1, inplace = True)

# We then check if all the columns are spelled correctly so as to allow 
# manipulation and merging with the other dataframes
df2.columns.values[3:5] = ['DW_A_NUMBER_INT','DW_B_NUMBER_INT']

# Then we proceed to check for null or missing values
# and return the sum of those values
df2.isnull().values.any()
df2.isnull().sum()

# We realize that again, the 'SITE_ID' column has missing values
df2 = df2.fillna('Not identified')

# Also check for duplicate values
df2.duplicated()
# This dataframe had one duplicate value which we proceed to drop
df2 = df2.drop_duplicates()

# Now we load our dataframe
df2
```

```python
# Cleaning the third telecom dataframe
# Again, we drop the columns which we won't use during analysis 
df3.drop('CELL_ON_SITE', axis = 1, inplace = True)
df3.drop('COUNTRY_A', axis = 1, inplace = True)
df3.drop('COUNTRY_B', axis = 1, inplace = True)

# Afterwards we spell check all our columns so as to match the preceding dataframes
df3.columns.values[5:] = ['CELL_ID','SITE_ID']

# We check for null or missing values in our dataframe
df3.isnull().values.any()
df3.isnull().sum()

# We substitute the null/missing values with 'Not Identified'
df3 = df3.fillna('Not Identified')

# We also check for duplicate values
print (df3.duplicated())
# This dataframe has duplicate values
df3.drop_duplicates()

# Now we load our dataframe
df3
```

## **3.2 Cleaning the geographical dataframe**


```python
# We also go ahead to clean the geographical dataframe
# Upon creating the dataframe, we drop any unwanted or null columns that are useless in our analysis
gdf1.drop('Unnamed: 0', axis=1, inplace = True)

# We check to see how our first five records look like
gdf1.head()

# Here we check to confirm if our data has any missing values 
gdf1.isnull().values.any()
gdf1.isnull().sum()

# We also check for duplicate values
gdf1.duplicated()

# After running the previous lines of code, we realize that
# 'status' and 'area' columns have some null/missing values
# We can then go ahead to replace those values with "Not Identified"
gdf1 = gdf1.fillna("Not identified")
gdf1
```

# **4. Merging the dataframes**

## **4.1 Merging the telecom dataframes**


```python
# Here we attempt to merge the telecom dataframes to create
# a combined dataframe for the three days 
TD = pd.concat([df1,df2,df3], axis = 0)

# We then confirm that the merged dataframe doesn't have any missing values
print (TD.isnull().values.any())

# We then load the merged dataframe
TD
```

## **4.2 Merging the geographical dataframe with the merged telecom dataframes**

```python
# Here we merge the combined telecom dataframe with the geographical dataframe
# then refer to it as the main dataframe
Main_DataFrame = pd.merge(TD, gdf1, how = 'inner', left_on = 'SITE_ID', right_on = 'SITE_CODE')
Main_DataFrame

# We then check if we have any all null or missing values
Main_DataFrame.isnull().any()
# We cleaned the data properly so there were no null values

# Afterwards, we check for any duplicate values and drop them so as to avoid redundancies
Main_DataFrame.duplicated()
Main_DataFrame = Main_DataFrame.drop_duplicates()
Main_DataFrame
```
